## Indexing

Akan dilakukan `indexing` pada file documents.csv. 

```
# jika menggunaakan Google Colab, mount data dari google drive

from google.colab import files
from google.colab import drive

drive.mount('/content/drive')
```

```
# TODO
# import dan tampilkan documents.csv dalam pandas dataframe

import pandas as pd
df = pd.read_csv('../documents.csv')
df.head(10)
```

```
# TODO
# Melakukan indexing pada file documents.csv (hint: gunakan IterDictIndexer)

iter_indexer = pt.IterDictIndexer("./index")
indexref = iter_indexer.index(df.to_dict(orient="records"))
```

```
# TODO
# Tampilkan statistik dari index yang sudah dibuat
index = pt.IndexFactory.of(indexref)
print(index.getCollectionStatistics().toString())
```

## Retrieval

Melakukan `querying` pada documents yang sudah di-index.


```
# TODO
# import dan tampilkan query dari query.csv dalam pandas dataframe
import pandas as pd
df = pd.read_csv('/../query14.csv')
df.head(10)
```

### BM25

```
# TODO
# Melakukan querying menggunakan scoring function BM25 dengan 
# hanya mengambil top 10 documents untuk setiap query
# Tampilkan hasil search
retr_BM25 = pt.BatchRetrieve(index, controls = {"wmodel": "BM25"})

retr_BM25.setControl("wmodel", "BM25")
retr_BM25.setControls({"wmodel": "BM25"})

res_BM25 = retr_BM25.transform(df)

res_BM25
```

**Menampilkan hasil search:**

Top 10 documents untuk setiap query

```
res_BM25_TOP10 = (retr_BM25 % 10).transform(df)
res_BM25_TOP10
```

```
# TODO
# Simpan hasil search query BM25 dengan penamaan file Tugas1_[NPM]_BM25.res
pt.io.write_results(res_BM25_TOP10, "../Tugas1_BM25.res")
```

### Dirichlet Language Model

```
# TODO
# Melakukan querying menggunakan scoring function 
# Dirichlet Language Model dengan hanya mengambil top 10 documents untuk setiap query
# Tampilkan hasil search seluruh query
retr_DLM = pt.BatchRetrieve(index, controls = {"wmodel": "DirichletLM"})

retr_DLM.setControl("wmodel", "DirichletLM")
retr_DLM.setControls({"wmodel": "DirichletLM"})

res_DLM = retr_DLM.transform(df)

res_DLM
```

Top 10 documents untuk setiap query:

```
res_DLM_TOP10 = (retr_DLM % 10).transform(df)
res_DLM_TOP10
```

```
# TODO
# Simpan hasil search query Dirichlet Language Model dengan penamaan file Tugas1_[NPM]_DLM.res
pt.io.write_results(res_DLM_TOP10, "../Tugas1_DLM.res")
```

## Evaluation

```
# TODO
# import qrels dan tampilkan. (hint: parse qrels dari local test collection)
qrels14 = pt.io.read_qrels("../TUGAS_1/qrels14")
qrels14
```

### BM25

```
# TODO
# Melakukan evaluasi hasil retrieval BM25 seluruh query 
# dengan metric evaluasi precision@10, recall@10, dan MRR
from pyterrier.measures import *
eval_BM25_all = pt.Experiment([res_BM25],
                              res_BM25,
                              qrels14,
                              names=["BM25"],
                              eval_metrics=["P_10", "recall_10", RR])
eval_BM25_all
```

```
# TODO
# Melakukan evaluasi hasil retrieval BM25 per query 
# dengan metric evaluasi precision@10, recall@10, dan MRR
eval_BM25_per_query = pt.Experiment([res_BM25],
                                    res_BM25,
                                    qrels14,
                                    names=["BM25"],
                                    eval_metrics=["P_10", "recall_10", RR],
                                    perquery=True)
eval_BM25_per_query
```

```
eval_BM25_per_query_pivot = eval_BM25_per_query.pivot(index=["name","qid"], 
                                                      columns="measure",
                                                      values="value")
eval_BM25_per_query_pivot
```

```
# TODO
# simpan hasil evaluasi dalam bentuk csv dengan row setiap query id 
# dan kolom setiap metrik evaluasi.
# Format penamaan file: Tugas1_[NPM]_eval_BM25.csv
eval_BM25_per_query_pivot.to_csv('../Tugas1_eval_BM25.csv')
```

### Dirichlet Language Model

```
# TODO
# Melakukan evaluasi hasil retrieval Dirichlet Language Model seluruh query dengan metric evaluasi precision@10, recall@10, dan MRR
from pyterrier.measures import *
eval_DLM_all = pt.Experiment([res_DLM],
                             res_DLM,
                             qrels14,
                             names=["DLM"],
                             eval_metrics=["P_10", "recall_10", RR])
eval_DLM_all
```

```
# TODO
# Melakukan evaluasi hasil retrieval Dirichlet Language Model per query dengan metrik evaluasi precision@10, recall@10, dan MRR
eval_DLM_per_query = pt.Experiment([res_DLM],
                                   res_DLM,
                                   qrels14,
                                   names=["DLM"],
                                   eval_metrics=["P_10", "recall_10", RR],
                                   perquery=True)
eval_DLM_per_query
```

```
eval_DLM_per_query_pivot = eval_DLM_per_query.pivot(index=["name","qid"], 
                                                    columns="measure",
                                                    values="value")
eval_DLM_per_query_pivot
```

```
# TODO
# simpan hasil evaluasi dalam bentuk csv dengan row setiap query id dan kolom setiap metrik evaluasi. Format penamaan file: Tugas1_[NPM]_eval_DLM.csv
eval_DLM_per_query_pivot.to_csv('../TUGAS_1/Tugas1_eval_DLM.csv')
```

## **Analisis Hasil**

####**1. Lakukan evaluasi hasil keseluruhan query. Manakah metode yang memiliki efektivitas lebih baik antara BM25 dan LM? Apakah perbedaan skor yang diperoleh sinifikan secara statistik?**

```
# Mendefinisikan data untuk dilakukan analisis lebih lanjut
data_retrieve_sign = pt.BatchRetrieve(index)
data_sign = data_retrieve_sign.transform(df)
data_sign
```

```
# dataframe unique qid dan query untuk kebutuhan merge data
# dengan tujuan membandingkan hasil evaluasi per-query dari BM25 dan DLM

data_1 = pd.DataFrame(data_sign, 
                      columns=['qid', 
                               'query'])
data_unique = data_1.drop_duplicates(subset = ["qid"])
data_unique = data_unique.reset_index().drop(["index"], axis=1)
data_unique
```

#### **Signifikansi Testing**


* Misal, metode BM25 akan dijadikan baseline dalam uji signifikansi yang akan dilakukan.

```
# Significance Testing
# Kita dapat melakukan pengujian signifikansi dengan 
# menentukan indeks transformator mana yang akan dijadikan sebagai baseline
eval_all_sign = pt.Experiment([res_BM25, res_DLM],
                              data_sign,
                              qrels14,
                              names=["BM25", "DLM"],
                              eval_metrics=["P_10", "recall_10", RR],
                              baseline=0,
                              correction='bonferroni')
eval_all_sign
```

* Berdasarkan uji signifikansi yang telah dilakukan di antara metode BM25 dan DLM, dapat disimpulkan bahwa: tidak adanya perbedaan skor yang signifikan secara statistik pada semua metrik evaluasi precision@10, recall@10, dan MRR (`P_10`, `recall_10` dan `RR`). Karena p-value dari semua metrik evaluasi > 0.05.

* Berdasarkan output uji signifikansi tersebut, juga dapat disimpulkan bahwa **metode DLM** meningkatkan nilai presisi_10 rata-rata untuk 2 kueri, meningkatkan nilai recall_10 rata-rata untuk 2 kueri dan juga meningkatkan nilai MRR untuk 2 kueri.

* Berdasarkan output uji signifikansi (secara keseluruhan) tersebut, juga dapat disimpulkan bahwa metode yang memiliki tingkat efektivitas yang lebih baik adalah **metode DLM**.
